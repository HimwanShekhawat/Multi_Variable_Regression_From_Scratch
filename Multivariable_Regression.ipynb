# ðŸ“˜ Multivariable Linear Regression
This notebook demonstrates multivariable linear regression from scratch using gradient descent. We generate synthetic data with three features and learn weights and bias through optimization.

## ðŸ”§ Generate Synthetic Training Data
We create training data with three features, a known linear relationship, and random noise.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Number of samples and features
num_samples = 100
num_features = 3

# Generate random x values (features)
x_train = 2 * np.random.rand(num_samples, num_features)

# Define true weights (slopes) and intercept
true_weights = np.array([[3], [2], [1]])  # shape: (3, 1)
true_intercept = 4

# Generate noise
noise = np.random.randn(num_samples, 1)

# Compute y values
y_train = true_intercept + x_train @ true_weights + noise  # @ is matrix multiplication

# Plotting using only first feature for visualization
plt.figure(figsize=(8, 5))
plt.scatter(x_train[:, 0], y_train, color="green", label="Training data (vs 1st feature)", alpha=0.7)
plt.title("Dummy Linear Regression Data (3 features)")
plt.xlabel("x_train[:, 0] (1st feature)")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.show()

print("Features shape is:", x_train.shape)
print("Target shape is:", y_train.shape)
```

## ðŸ§® Implement Gradient Descent
We optimize the weights and bias to minimize the Mean Squared Error (MSE) between predicted and true values.

```python
# Initialize weights and bias
w = np.array([[2], [8], [7]])  # Initial guess for weights (shape: 3x1)
b = 0                          # Initial bias
alpha = 0.1                    # Learning rate
all_cost = []                  # To track cost function value at each iteration

# Perform 500 iterations of gradient descent
for _ in range(500):
    y_ = x_train @ w + b  # Predicted output (shape: 100x1)

    # Compute Mean Squared Error cost
    cost = (1 / (2 * x_train.shape[0])) * np.sum((y_ - y_train)**2)
    all_cost.append(cost)

    # Update weights and bias using gradients
    w = w - (alpha / x_train.shape[0]) * (x_train.T @ (y_ - y_train))
    b = b - (alpha / x_train.shape[0]) * np.sum(y_ - y_train)

# Plot training data again for reference (only first feature shown)
plt.figure(figsize=(8, 5))
plt.scatter(x_train[:, 0], y_train, color="green", label="Training data (vs 1st feature)", alpha=0.7)
plt.title("Dummy Linear Regression Data (3 features)")
plt.xlabel("x_train[:, 0] (1st feature)")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.show()

# Plot the cost function over iterations
plt.plot(all_cost)
plt.title("Cost over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.grid(True)
plt.show()
```
